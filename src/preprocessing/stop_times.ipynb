```python
all_times_df = pd.read_csv("../../Datasets/archived_feeds.txt")
all_times_df = all_times_df.drop(['archive_note'], axis=1)
all_times_df['feed_start_date'] = pd.to_datetime(all_times_df['feed_start_date'], format='%Y%m%d')
all_times_df['feed_end_date'] = pd.to_datetime(all_times_df['feed_end_date'], format='%Y%m%d')
all_times_df
```

```python
# obtain all final events data for 2022-2026
df_list = []
for i in range(2, 6):
    dir = glob.glob(f'/Users/sierrawelsch/CS4100_Final/Datasets/202{i}/*.csv')
    for month_csv_file in dir:
        month_csv_file_df = pd.read_csv(month_csv_file)
        
        df_list.append(month_csv_file_df)
print(len(df_list))
final_events_df = pd.concat(df_list, ignore_index=True)
final_events_df
```

```python
new_final_events_df = final_events_df.drop(["vehicle_id", "vehicle_label", "direction_id", "parent_station", "trunk_route_id", "direction", "stop_name"], axis = 1)
new_final_events_df['stop_sequence_id'] = new_final_events_df['stop_sequence'].combine_first(new_final_events_df['sync_stop_sequence'])
new_final_events_df = new_final_events_df.drop(['sync_stop_sequence', 'stop_sequence'], axis=1)
new_final_events_df = new_final_events_df.dropna()

new_final_events_df
```

```python
# merging the tables to map each stop times zip file to the appropriate event based on the event date (service_date)
new_final_events_df['service_date'] = pd.to_datetime(new_final_events_df['service_date']).astype('datetime64[ns]')
all_times_df['feed_start_date'] = pd.to_datetime(all_times_df['feed_start_date']).astype('datetime64[ns]')

new_final_events_df = new_final_events_df.sort_values('service_date')
all_times_df = all_times_df.sort_values('feed_start_date')

merged = pd.merge_asof(
    new_final_events_df,
    all_times_df,
    left_on='service_date',
    right_on='feed_start_date',
    direction='backward'
)
merged = merged.sort_values('service_date')
merged
```

```python
import os
import requests
import zipfile

os.makedirs("stop_times", exist_ok=True)
os.makedirs("zips", exist_ok=True)

# Keep track of currently downloaded ZIPs
downloaded_zips = {}  # url -> local path

def download_zip(url):
    filename = url.split("/")[-1]
    path = os.path.join("zips", filename)
    
    if url not in downloaded_zips:
        if not os.path.exists(path):
            print(f"Downloading {filename}")
            r = requests.get(url)
            with open(path, "wb") as f:
                f.write(r.content)
        downloaded_zips[url] = path
    return downloaded_zips[url]

def extract_stop_times(zip_path):
    filename = zip_path.split("/")[-1].replace(".zip", ".txt")
    output_path = os.path.join("stop_times", filename)
    
    if not os.path.exists(output_path):
        with zipfile.ZipFile(zip_path, 'r') as zip_ref:
            zip_ref.extract("stop_times.txt", "temp_extract")
            os.rename("temp_extract/stop_times.txt", output_path)
            os.rmdir("temp_extract")
    return output_path

def clean_old_zips(zip_path):
    """Delete ZIPs not in the active set."""
    os.remove(f'src/preprocessing/{zip_path}')
    print(f"Deleted old ZIP: {zip_path}")
```
