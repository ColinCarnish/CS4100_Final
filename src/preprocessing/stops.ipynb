```python
import pandas as pd
import json
```

```python
df = pd.read_csv("../../Datasets/final_data.csv")
route_to_stops = (
    df.sort_values("stop_sequence")
      .groupby("route_id")["stop_name"]
      .apply(lambda x: list(dict.fromkeys(x)))
      .to_dict()
)

with open("../../Datasets/stops_on_route.json", "w") as f:
    json.dump(route_to_stops, f, indent=2)
```
